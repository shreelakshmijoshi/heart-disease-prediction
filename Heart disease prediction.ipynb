## Data Visualization

```python
plt.rcParams['figure.figsize'] = (10,15)
sns.pairplot(df)
```

```python
# checking the distribution of age amonng the patients

plt.rcParams['figure.figsize'] = (10, 7)
sns.distplot(df['age'])
plt.title('Distribution of Age')
```

```python
# checking the distribution of gender among the patients
df['sex'].value_counts()
```

```python
df.hist(figsize=(15,15))
```

```python
# relation between age and num

sns.swarmplot(df['num'], df['age'], palette = 'dark')
plt.title('Relation of Age and num', fontsize = 10)
```

```python
# relation between cholestrol and num

sns.swarmplot(df['num'], df['chol'], palette = 'dark')
plt.title('Relation of chol and num', fontsize = 10)
```
